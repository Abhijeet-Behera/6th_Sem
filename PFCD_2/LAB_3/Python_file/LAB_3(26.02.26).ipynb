```python
import pandas as pd
import matplotlib.pyplot as plt

celsius = lambda f:5/9*(f-32)
temps =[(f,celsius(f)) for f in range (0,101,10)]#creating list of tupples for 0 to 100 F in 10 degree increment
temps_df = pd.DataFrame(temps, columns=['Fahrenheit','Celsius'])#Dataframe
axes = temps_df.plot(x='Fahrenheit',y='Celsius', style='.-', legend=True)#plotting the datapoints
axes.set_ylabel('Celsius')#setting y-axis label
plt.show()
```

```python
from collections import Counter
import matplotlib.pyplot as plt

grades = [83, 95, 91, 87, 70, 0, 85, 100, 67, 73, 77, 0]
histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)

plt.bar([x+5 for x in histogram.keys()], histogram.values(),width=10, edgecolor=(0, 0, 0))
plt.axis([-5, 105, 0, 5])
plt.xticks(range(0, 101, 10))
plt.xlabel("Decile")
plt.ylabel("# of Students")
plt.title("Grade Distribution")
plt.show()
```

```python

#Inheritance using getter and setter methods
#question from th BOOK

from decimal import Decimal


class CommissionEmployee:
    """An employee who gets paid commission based on gross sales."""

    def __init__(self, first_name, last_name, ssn, gross_sales, commission_rate):
        """Initialize CommissionEmployee's attributes."""
        self._first_name = first_name
        self._last_name = last_name
        self._ssn = ssn
        self.gross_sales = gross_sales          # Validate via property
        self.commission_rate = commission_rate  # Validate via property

    @property
    def first_name(self):
        return self._first_name

    @property
    def last_name(self):
        return self._last_name

    @property
    def ssn(self):
        return self._ssn

    @property
    def gross_sales(self):
        return self._gross_sales

    @gross_sales.setter
    def gross_sales(self, sales):
        if sales < Decimal('0.00'):
            raise ValueError("Gross sales must be >= 0.0")
        self._gross_sales = sales

    @property
    def commission_rate(self):
        return self._commission_rate

    @commission_rate.setter
    def commission_rate(self, rate):
        if rate <= Decimal('0.0') or rate >= Decimal('1.0'):
            raise ValueError("Commission rate must be > 0.0 and < 1.0")
        self._commission_rate = rate

    def earnings(self):
        """Calculate earnings."""
        return self.gross_sales * self.commission_rate

    def __repr__(self):
        return (f'CommissionEmployee: {self.first_name}{self.last_name}\n'
                f'Social Security Number: {self.ssn}\n'
                f'Gross sales:{self.gross_sales:.2f}\n'
                f'commission Rate:{self.commission_rate:.2f}')
    
class SalariedCommissionEmployee(CommissionEmployee):
    def __init__(self, first_name, last_name, ssn, gross_sales, commission_rate,base_salary):
        super().__init__(self, first_name, last_name, ssn, gross_sales, commission_rate)
        self.base_salary=base_salary
        
    @property
    def base_salary(self):
        return self.base_salary
    
    @base_salary.setter
    def base_salary(self,salary):
        if salary < Decimal('0.00'):
            raise valueError('Base salary must be >=0')
        self._base_salary=salary
    
```
