```python
#Q1) What is the significance of classes in Python programming,
#and how do they contribute to objectoriented programming?

'''  Classes in Python serve as the blueprints for creating objects, acting as the fundamental building blocks 
of Object-Oriented Programming (OOP). Their primary significance lies in encapsulation, where they bundle data 
(attributes) and behaviors (methods) into a single unit, making code more modular, readable, and reusable. 
By using classes, developers can implement inheritance, allowing new classes to adopt the properties of existing 
ones, and polymorphism, which enables different objects to be treated through a uniform interface. 
This structure promotes the "Don't Repeat Yourself" (DRY) principle, as it allows for the creation of complex, 
scalable systems where real-world entities are modeled accurately. Ultimately, classes transform code from 
a series of linear instructions into a collaborative ecosystem of independent objects, significantly reducing 
logical errors and simplifying long-term maintenance.'''
```

```python
#Q2)Create a custom Python class for managing a bank account with basic functionalities like deposit and withdrawal?
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self._balance = balance 
    @property
    def balance(self):
        return self._balance
    @balance.setter
    def balance(self, value):
        if value < 0:
            print("Error: Balance cannot be negative.")
        else:
            self._balance = value

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount  
        return self.balance

    def withdraw(self, amount):
        if 0 < amount <= self._balance:
            self.balance -= amount  
        else:
            print("low balance.")
        return self.balance


acc = BankAccount("Alex", 100)
acc.deposit(50)
print(acc.balance)
acc.withdraw(20)
print(acc.balance)
acc.withdraw(200)
print(acc.balance)
```

```python
#Q3
class Book:
    def __init__(self, title):
        self.title = title
        self.chapters = [] 

    def add_chapter(self, title, pages):
        self.chapters.append({"title": title, "pages": pages})

    def total_pages(self):
        return sum(map(lambda c: c["pages"], self.chapters))


my_book = Book("Python Basics")
my_book.add_chapter("Introduction", 10)
my_book.add_chapter("Variables", 25)
my_book.add_chapter("Functions", 35)

print(f"Book: {my_book.title}")
print(f"Total Pages: {my_book.total_pages()}")
```

```python
#Q5)
from datetime import datetime

class Time:
    def convert_time(self, time_input):
        try:
            t = datetime.strptime(time_input, "%H:%M:%S")
            return f"Valid time: {t.strftime('%I:%M:%S %p')}"
        except ValueError:
            return "Error: Invalid format or values. Use HH:MM:SS (00-23 hours)."


t_obj = Time()
user_time = input("Enter time in 24-hour format (HH:MM:SS): ")
print(t_obj.convert_time(user_time))
```
