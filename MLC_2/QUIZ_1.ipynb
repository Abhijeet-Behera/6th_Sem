# Q1

```python
import numpy as np
import pandas as pd

# Generate dataset (8 rows, 2 columns)
data = np.random.randint(1, 20, size=(8,2))

# Convert to DataFrame for better visualization
df = pd.DataFrame(data, columns=['X', 'Y'])

print("Dataset (8x2):")
print(df)
```

```python
import matplotlib.pyplot as plt

# Scatter plot
plt.scatter(df['X'], df['Y'])

plt.title("Scatter Plot of Data")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

```python
from sklearn.cluster import KMeans

# Apply KMeans for k = 1 to 6
for k in range(1,7):
    
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(df)
    
    print(f"K = {k}")
    print("Cluster Centers:")
    print(kmeans.cluster_centers_)
    print()
```

```python
inertia_values = []

for k in range(1,7):
    
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(df)
    
    inertia_values.append(kmeans.inertia_)

print("Inertia values for each K:")
print(inertia_values)
```

```python
k_values = range(1,7)

plt.plot(k_values, inertia_values, marker='o')

plt.title("Elbow Method for Optimal K")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia (Sum of Squared Distances)")

plt.xticks(k_values)

plt.show()
```

```python
import numpy as np

# Step Activation Function
def step_function(x):
    if x >= 0:
        return 1
    else:
        return 0


# ReLU Activation Function
def relu_function(x):
    return max(0, x)


# Hyperbolic Tangent Activation Function
def tanh_function(x):
    return np.tanh(x)


# Input values
x_values = [-2, -1, 0, 1, 2]

print("Input Values:", x_values)
print("\nActivation Function Outputs:\n")

for x in x_values:
    print("Input:", x)
    print("Step Output:", step_function(x))
    print("ReLU Output:", relu_function(x))
    print("Tanh Output:", tanh_function(x))
    print("---------------------------")
```

```python
# Input values
x_values = [-2, -1, 0, 1, 2]

print("x\tStep\tReLU\tTanh")

for x in x_values:
    step = step_function(x)
    relu = relu_function(x)
    tanh = tanh_function(x)

    print(f"{x}\t{step}\t{relu}\t{tanh:.4f}")
```

```python
# =========================================
# Import Required Libraries
# =========================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


# =========================================
# (a) Generate a data sheet of shape (8,2)
# =========================================
print("\n===== (a) Generate Dataset of Shape (8,2) =====")

data = np.random.randint(1, 20, size=(8,2))
df = pd.DataFrame(data, columns=['X', 'Y'])

print("Generated Dataset:")
print(df)


# =========================================
# (b) Draw Scatter Plot to Visualize Data
# =========================================
print("\n===== (b) Scatter Plot of Data =====")

plt.scatter(df['X'], df['Y'])
plt.title("Scatter Plot of Dataset")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()


# =========================================
# (c) Apply K-Means for different K values
#     K = 1 to 6
# =========================================
print("\n===== (c) Apply K-Means for K = 1 to 6 =====")

inertia_values = []

for k in range(1,7):
    
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(df)
    
    print("\nK =", k)
    print("Cluster Centers:")
    print(kmeans.cluster_centers_)
    
    # store inertia for elbow method
    inertia_values.append(kmeans.inertia_)


# =========================================
# (d) Calculate Sum of Squared Distance
#     (Inertia)
# =========================================
print("\n===== (d) Sum of Squared Distance (Inertia) =====")

for i, val in enumerate(inertia_values, start=1):
    print("K =", i, " Inertia =", val)


# =========================================
# (e) Determine Optimal K using Elbow Graph
# =========================================
print("\n===== (e) Elbow Method Graph =====")

k_values = range(1,7)

plt.plot(k_values, inertia_values, marker='o')
plt.title("Elbow Method for Optimal K")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia (Sum of Squared Distance)")
plt.xticks(k_values)

plt.show()
```
